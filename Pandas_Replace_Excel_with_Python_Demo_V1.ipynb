## Data Engineering

```python
# df head
df.sample(10)
```

Lets extract Cabin Name from Cabin

```python
# function to extract Cabin names from Cabin
import re
def extract_cabin(name):
    try:
        return re.search(r'[A-Za-z]+',name).group(0)
    except:
        return "Missing"
```

```python
# create new column cabin name by mapping
df["Cabin_Name"] = df.Cabin.apply(extract_cabin)
```

```python
# df sample 5
df.groupby("Cabin_Name")["Survived"].mean().plot(kind='bar')
```

Lets also extract titles from Name

```python
# function to extract titles from names
import re
def extract_title(name):
    try:
        return re.search(r'[A-Za-z]+\.',name).group(0)
    except:
        return "Missing"
```

```python
# Create new column title by mapping
df["title"] = df.Name.apply(extract_title)
```

```python
# df sample 5
df.groupby("title")["Survived"].mean().plot(kind='bar')
```

## Machine Learning

```python
# import Random Forest , cross val score

from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
```

```python
# df head
df.head()
```

```python
# define X and y
X = df.drop(["PassengerId","Survived","Name","Ticket","Cabin"],axis=1)

y = df["Survived"]
```

```python
#X = get dummies 
X = pd.get_dummies(X)
```

```python
#instantiate RF
rf = RandomForestClassifier()
```

```python
#cross val score
cross_val_score(rf,X,y,cv=10).mean()
```

```python
#fit RF
rf.fit(X,y)
```

```python
#Look at feature importance
sorted(zip(rf.feature_importances_,X.columns))[::-1]
```
